#### Plot lysimeter vs calculated

```python
from sklearn.linear_model import LinearRegression
def scatter(obs, sim, lim=10):
    obs[obs>lim] = np.nan
    sim[sim>lim] = np.nan
    compare = pd.DataFrame(data={"obs":obs, "sim":sim}).dropna()
    obs = compare["obs"].to_numpy()  
    sim = compare["sim"].to_numpy()
    
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(9,9))
    ax.scatter(obs, sim, marker=".", c="k")
    maxv = round(max(obs.max(), sim.max()),0)
    ax.plot(np.arange(0,maxv+1), np.arange(0,maxv+1), c="r")
    ax.grid(True, linestyle="-.", alpha=1)
    ax.set_xlim(0, maxv)
    ax.set_ylim(0, maxv)
    model = LinearRegression(fit_intercept=True)
    x1=obs.reshape(-1, 1)
    y1=sim.reshape(-1, 1)
    model.fit(x1, y1)
    xfit = obs.reshape(-1, 1)
    yfit = model.predict(xfit)
    ax.plot(xfit, yfit)
    plt.tight_layout()
```

```python
obs = et_lysid["GS-2"].to_numpy()
sim = veronikaetd["ET_pm"].to_numpy()

scatter(obs, sim)
```
